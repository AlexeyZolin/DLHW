```python
import torch
#Для работы с тремя признаками необходимо изменить размерность weights на (1, 3).
#Задача предсказания погоды, признаки: влажность, скорость ветра, вероятность осадков.
Monday = torch.tensor([[0.6, 0.6, 0.2]])
Tuesday = torch.tensor([[0.2, 0.5, 0.3]])
Wednesday = torch.tensor([[0.5, 0.5, 0.1]])
Thursday = torch.tensor([[0.3, 0.5, 0.4]])
Friday = torch.tensor([[0.2, 0.2, 0.1]])


dataset = [
    (Monday, torch.tensor([[0.5]])),
    (Tuesday, torch.tensor([[0.4]])),
    (Wednesday, torch.tensor([[0.6]])),
    (Thursday, torch.tensor([[0.4]])),
    (Friday, torch.tensor([[0.7]])),

]

torch.manual_seed(2023)

weights = torch.rand((1, 3), requires_grad=True)
bias = torch.rand((1, 1), requires_grad=True)

mse_loss_fn = torch.nn.MSELoss()

optimizer = torch.optim.SGD([weights, bias], lr=1e-5)

def predict_threat_score(obj: torch.Tensor) -> torch.Tensor:
    return obj @ weights.T + bias

def calc_loss(predicted_value: torch.Tensor, ground_truth: torch.Tensor) -> torch.Tensor:
    return mse_loss_fn(predicted_value, ground_truth)

num_epochs = 15

for i in range(num_epochs):
    for x, y in dataset:
        optimizer.zero_grad()
        threat_score = predict_threat_score(x)

        loss = calc_loss(threat_score, y)
        loss.backward()
        print(loss)
        optimizer.step()
```
